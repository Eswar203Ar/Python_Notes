### Identity Operators

1.**is**

2.**is not**

- Identity operators used to check if two variables point to the same object in memory
- **is** - Returns **True** if  two variables point to the same object in memory
- **is not** - Returns **True** if two variables do not point to same object in memory

Differnce  from == :
- **==** --> check for values(content)
-  **is** --> checks identity(memory location)

The difference between == and is is mainly what they compare.

1. == Operator

- Used to compare values.

- It checks whether two variables have the same value.

2.  is Operator

- Used to compare memory location (object identity).

- It checks whether two variables refer to the same object in memory

```python
# when we use else with for or while

# else can be used with for and while loops in Python.

# The else block runs when the loop finishes normally.

# If the loop stops because of a break statement, the else block will NOT execute.

# The else block usually contains code that should run after successful completion of the loop.

# It is commonly used in searching or checking conditions (for example: prime number checking)

nums = [3,7,4,9,8,2]
ele = 5
for x in nums:
    if ele == x:
        print('found')
else:
    print('Not found')
```

#### why function need arguments ?

- Arguments are used to pass values to a function so the function can work with different data.

- Example

- Required
- Default

#### Define Default arguments functions ?

**Default argument:**
- A parameter that has a default value if no argument is provided.

- Example

def greet(name="Guest"):
    print("Hello", name)

greet()
greet("Ram")

#### Define Keyword arguments function ?

**Keyword arguments:**
- Arguments passed using parameter names.

- Example

def student(name, age):
    print(name, age)

student(age=20, name="Ravi")


#### Define Variable arguments function ?

**Variable arguments:**
- A function that can accept any number of arguments.

**Uses *.**

- Example

def add(*numbers):
    print(sum(numbers))

add(1,2)
