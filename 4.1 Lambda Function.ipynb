Lambda function in python
Lambda function are small anonymous functions defined using lambda keyword.They can hava any number of arguments,but only one expression.They commonly used for short operations or as arguments to higher order-function.


```python
## Syntax 
lambda arguments : expression
```

```python
def even(num):
    if num%2 == 0:
        return True
even(24)
```

```python
even1 = lambda num : num%2==0
even1(10)
```

```python
def addition(a,b):
    return a+b
addition(3,4)
```

```python
addition1 = lambda a,b : a+b
addition(4,5)
```

```python
numbers = [1,2,3,4,5]
list(map(lambda x: x**2,numbers))
```

```python
addition2 =  lambda x,y,z : x+y+z
addition2(2,5,7)
```
