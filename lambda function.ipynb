Lambda function in python
lambda function is small anonymous function defined using the lambda keyword. They can have any number of arguments but only one expression. They are commonly used for short operations or as arguments to higher order function

```python
# Syntax
lambda aruguments:expression
```

```python
# addition operation in functions
def addition(a,b):
    return a+b
print(addition(3,4))
```

```python
# Lambda function'
addition = lambda a,b:a+b
type(addition)
print(addition(2,3))
```

```python
# lambda function even number
even1 = lambda num : num%2==0
even1(12)
```

```python
# function 
def even(num):
    if num%2==0:
        return True
even(24)
```

```python
# lambda addition
addition1 = lambda x,y,z : x+y+z
print(addition1(2,3,4))
```

```python
# map() - applies a map function to all items to alist
numbers = [1,2,3,4,5,6]
def square(number):
    return number ** 2
square(2)
```

```python
# lambda x : x**2
# map(lambda x : x**2,numbers)
list(map(lambda x : x**2,numbers))
```
