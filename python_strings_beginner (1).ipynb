# Python Strings — 30 Beginner Programs


---
## 1. Print a String

```python
name = "Hello, Python!"
print(name)
```

---
## 2. Find Length of a String

```python
text = "Python"
print("Length:", len(text))
```

---
## 3. Convert to Uppercase and Lowercase

```python
text = "Hello World"
print("Uppercase:", text.upper())
print("Lowercase:", text.lower())
```

---
## 4. Reverse a String

```python
text = "Python"
reversed_text = text[::-1]
print("Original:", text)
print("Reversed:", reversed_text)
```

---
## 5. Access Characters by Index

```python
text = "Python"
print("First character :", text[0])
print("Last character  :", text[-1])
print("Third character :", text[2])
```

---
## 6. Slice a String

```python
text = "Hello, Python!"
print("First 5 chars :", text[:5])
print("Last 7 chars  :", text[-7:])
print("Middle slice  :", text[7:13])
```

---
## 7. Concatenate Two Strings

```python
first = "Hello"
second = "World"
result = first + ", " + second + "!"
print(result)
```

---
## 8. Repeat a String

```python
text = "Ha"
print(text * 5)
```

---
## 9. Check if a Substring Exists

```python
sentence = "I love Python programming"
print("'Python' in sentence:", "Python" in sentence)
print("'Java' in sentence  :", "Java" in sentence)
```

---
## 10. Count Occurrences of a Character

```python
text = "banana"
print("Count of 'a':", text.count('a'))
print("Count of 'n':", text.count('n'))
```

---
## 11. Replace a Word in a String

```python
text = "I love Java programming"
new_text = text.replace("Java", "Python")
print("Before:", text)
print("After :", new_text)
```

---
## 12. Strip Whitespace

```python
text = "   Hello, Python!   "
print("Original :", repr(text))
print("Stripped :", repr(text.strip()))
print("L-Strip  :", repr(text.lstrip()))
print("R-Strip  :", repr(text.rstrip()))
```

---
## 13. Split a String into a List

```python
sentence = "Python is easy to learn"
words = sentence.split()
print("Words:", words)
print("Number of words:", len(words))
```

---
## 14. Join a List into a String

```python
words = ["Python", "is", "fun"]
sentence = " ".join(words)
print("Joined:", sentence)

csv_line = ",".join(words)
print("CSV   :", csv_line)
```

---
## 15. Check Start and End of a String

```python
text = "Hello, Python!"
print("Starts with 'Hello':", text.startswith("Hello"))
print("Ends with 'Python!':", text.endswith("Python!"))
print("Starts with 'World':", text.startswith("World"))
```

---
## 16. Find Position of a Substring

```python
text = "Hello, Python!"
print("Position of 'Python':", text.find("Python"))
print("Position of 'Java'  :", text.find("Java"))  # -1 means not found
```

---
## 17. Check String Type (isalpha, isdigit, isalnum)

```python
s1 = "Hello"
s2 = "12345"
s3 = "Hello123"

print(f"'{s1}' -> isalpha: {s1.isalpha()}, isdigit: {s1.isdigit()}, isalnum: {s1.isalnum()}")
print(f"'{s2}' -> isalpha: {s2.isalpha()}, isdigit: {s2.isdigit()}, isalnum: {s2.isalnum()}")
print(f"'{s3}' -> isalpha: {s3.isalpha()}, isdigit: {s3.isdigit()}, isalnum: {s3.isalnum()}")
```

---
## 18. Capitalize and Swap Case

```python
text = "hello world"
print("capitalize() :", text.capitalize())
print("title()      :", text.title())
print("swapcase()   :", text.swapcase())
```

---
## 19. Print Each Character of a String

```python
text = "Python"
for ch in text:
    print(ch)
```

---
## 20. Print Characters with Index

```python
text = "Python"
for i, ch in enumerate(text):
    print(f"Index {i}: {ch}")
```

---
## 21. Count Spaces in a String

```python
text = "Hello World from Python"
spaces = text.count(' ')
print(f"Text   : '{text}'")
print(f"Spaces : {spaces}")
```

---
## 22. Check if String is Empty

```python
s1 = ""
s2 = "hello"
s3 = "   "

print(f"'{s1}' is empty:", len(s1) == 0)
print(f"'{s2}' is empty:", len(s2) == 0)
print(f"'{s3}' is blank (only spaces):", s3.strip() == "")
```

---
## 23. Convert String to List of Characters

```python
text = "Python"
char_list = list(text)
print("Characters:", char_list)
```

---
## 24. Get ASCII Value of a Character

```python
chars = ['A', 'a', 'Z', 'z', '0', '9']
for c in chars:
    print(f"'{c}' -> ASCII: {ord(c)}")
```

---
## 25. Check if a String Contains Only Spaces

```python
s1 = "     "
s2 = "  hi  "
print(f"'{s1}' isspace:", s1.isspace())
print(f"'{s2}' isspace:", s2.isspace())
```

---
## 26. Format a String with f-string

```python
name = "Alice"
age = 25
city = "Hyderabad"
print(f"My name is {name}, I am {age} years old, and I live in {city}.")
```

---
## 27. Repeat Each Character in a String

```python
text = "Hello"
result = ''.join(c * 2 for c in text)
print("Original:", text)
print("Doubled :", result)
```

---
## 28. Remove a Specific Character from a String

```python
text = "h-e-l-l-o"
result = text.replace("-", "")
print("Original:", text)
print("Cleaned :", result)
```

---
## 29. Check if All Characters are Uppercase or Lowercase

```python
s1 = "HELLO"
s2 = "hello"
s3 = "Hello"

for s in [s1, s2, s3]:
    print(f"'{s}' -> isupper: {s.isupper()}, islower: {s.islower()}")
```

---
## 30. Count Words in a Sentence

```python
sentence = "Python is a great programming language"
word_count = len(sentence.split())
print("Sentence   :", sentence)
print("Word Count :", word_count)
```
