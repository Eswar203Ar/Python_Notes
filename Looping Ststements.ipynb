```python
# while loop
count = 0
while count < 5:
    print(count)
    count = count + 1    
```

```python
count = 0
while count % 2 == 0:
    print(count)
    count = count + 1
```

```python
# Loop control statements
# break
for i in range(10):
    if i == 5:
        break
    print(i)
        
```

```python
# continue
for i in range(10):
    if i%2 == 0:
        continue
    print(i)
```

```python
# pass
for i in range(5):
    if i == 3:
        pass
    print(i)
```

```python
# Nested for loop
for i in range(3):
    for j in range(2):
        print(f"i:{i} and j:{j}")
```

```python
# Example program'
# calculate the sum of n natural number using while loop and for loop
n = 10
sum = 0
count = 1
while count <= n:
    sum = sum + count
    count = count + 1
print("sum of the first n natural numbers", sum)
```

```python
# N natural number
n = 10
sum = 0
for i in range(11):
    sum = sum + i
print(sum)
```

```python
# prime number 1 to 100
for num in range(1,101):
    if num >1:
        for i in range(2,num):
            if num%i==0:
                break
        else:
            print(num)
```
