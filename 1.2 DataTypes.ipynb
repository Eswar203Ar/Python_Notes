```python
#String datatype
name = "Eswar"
print(name)
type(name)
```

```python
# string concatenate
name  = "Eswar" + str(5)
print(name)
type(name)
```

```python
## simple caculator
num1 = float(input("Enter the first number"))
num2 = float(input("Enter the second number"))

sum = num1 + num2
sub = num1 - num2
mult = num1 * num2
div = num1 / num2
mod = num1 % num2

print(sum)
print(sub)
print(mult)
print(div)
print(mod)
```
