Right-angled triangle

```python
n = 5
for i in range(1, n+1):
    print("*" * i)
```

Right angle triangle with numbers

```python
n = 5
for i in range(1, n+1):
    print("*" * i)
```

Inverted right angle triangle

```python
n = 5
for i in range(n, 0, -1):
    print("*" * i)
```

Right aligned triangle

```python
n = 5
for i in range(1, n+1):
    print(" " * (n-1) + "*" * i)
```

```python
#Number right angle triangle

    
n = 5
for i in range(1, n+1):
    for j in range(1, i+1):
        print(j, end = " ")
    print()
    
n = 5
for i in range(n, 0, -1):
    for j in range(1, i+1):
        print(j, end = " ")
    print()
```

```python
# Diamond pattren
n = 5
for i in range(1, n+1):
    print(" " * (n-1) + "*" * (2*i-1))
    
for i in range(n-1, 0, -1):
    print(" " * (n-1) + "*" * (2*i-1))
```
