## List comprehension
 1 . Basic Syntax:
[Expression for item in iterable with conditional logic]

2 . With conditional logic
[Expression for item in iterable if condition]

3 . Nested list compehension
[Expression for item1 in iterable1 for item2 in iterable2]

```python
## Basic Compehension
square = [x**2 for x in range(10)]
print(square)
```

```python
## List Compehension with condition
list = []
for i in range(10):
    if i%2==0:
        list.append(i)
print(list)        
```

```python
even_numbers = [num for num in range(10) if num%2==0]
print(even_numbers)
```

```python
odd_numbers = [num  for num in range(15) if num%2 != 0]
print(odd_numbers)
```

```python
## Nested list compehension
list1 = [1,2,3,4]
list2 = ['a','b','c','d']
pair = [[i,j] for i in list1 for j in list2]
print(pair)
```

```python
num1 = [12,13,14,15,16]
list2 = ['an','bn','cn','dn','en']
pair = [[i,j] for i in num1 for j in list2]
print(pair)
```
