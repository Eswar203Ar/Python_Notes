Sets

sets are a built in data type in python used to store collections of unique items. They are unorred, meaning that the elements do not follow a specific order, and they do not allow duplicates entries, and performing mathematical set operations like union, intersection,fiffernce and symmetric differnce


```python
set = {1,2,3,4,5}
print(set)
print(type(set))
```

```python
set = {}
print(type(set))
```

```python
# set do not allow the duplicate values
my_set = {1,2,3,3,4,5,6}
print(my_set)
```

```python
## Basic set operation 
# Adding and removing elements
my_set.add(7)
print(my_set)

## removing the element
my_set.remove(3)
print(my_set)
```

```python
## Key error in set
my_set.remove(10)
```

```python
my_set.discard(11)
print(my_set)
```

```python
## set methods
# pop method 
# removing the first element
removed_Element = my_set.pop()
print(removed_Element)
print(my_set)
```

```python
## clear all elements 
## clear method
my_set.clear()
print(my_set)
```

```python
## set membership test
my_set = {1,2,3,4,5,6}
print(3 in my_set)
print(10 in my_set)
```

```python
## mathametical operation of set
set1 = {1,2,3,4,5,6}
set2 = {4,5,6,7,8,9}

## union
# combing everything in the both sets
union_set = set1.union(set2)
print(union_set)
```

```python
## Intersection 
# only common elements for the both the sets
intersection = set1.intersection(set2)
print(intersection)
```

```python
## intersection update is used for update the set
set1.intersection_update(set2)
print(set1)
```

```python
## Differnce in set
# differnt element in set1
set1 = {1,2,3,4,5,6}
set2 = {4,5,6,7,8,9}

print(set1.difference(set2))
print(set2.difference(set1))
```

```python
## symmentric Differnce
## common elements removed in both the set2
set1.symmetric_difference(set2)
set2.symmetric_difference(set1)
```
