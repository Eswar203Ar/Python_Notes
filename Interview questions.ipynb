Basic interview questions

```python
# Reverse a string\
s = "hello"
print(s[::-1])
```

```python
#  Check Palindrome (String or Number)
s = "madam"
print(s == s[::-1])
```

```python
# Fibonacci Series
n = 10
a, b = 0, 1
for _ in range(n):
    print(a, end=" ")
    a, b = b, a + b
```

```python
#Factorial of a Number
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)
print(factorial(5))
```

```python
# Prime Number Check
n = 29
is_prime = all(n % i != 0 for i in range(2, int(n**0.5)+1)) if n > 1 else False
print(is_prime)
```

```python
# Find Largest of 3 Numbers
a, b, c = 10, 20, 15
print(max(a, b, c))
```

```python
 #Swap Two Variables Without Temp Variable
a, b = 5, 10
a, b = b, a
print(a, b)
```

```python
# Count Vowels in
# a String
s = "Interview"
vowels = "aeiouAEIOU"
count = sum(1 for char in s if char in vowels)
print(count)
```

```python
#  Armstrong Number (e.g., 153 = 1³ + 5³ + 3³)
n = 153
digits = [int(d) for d in str(n)]
print(sum(d**len(digits) for d in digits) == n)
```

```python
# Pattern Printing (e.g., Pyramid, Diamond)
n = 5
for i in range(1, n+1):
    print(" " * (n - i) + "*" * (2*i - 1))
    
for i in range(1, n+1):
    print(" " * (n - i) + "*" * (2*i - 1))
# Lower part
for i in range(n-1, 0, -1):
    print(" " * (n - i) + "*" * (2*i - 1))
```
