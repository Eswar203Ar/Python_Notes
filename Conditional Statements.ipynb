```python
# if statement
age = 18
if age>=18:
    print("You are eligible for vote")
```

```python
#else statement

age = 17
if age > 18:
    print("You are eligible for vote")
else:
    print("you are a minor")
```

```python
#elif statement
age = 12
if age < 13:
    print("you are minor")
elif age < 18:
    print("you are teenager")
else:
    print("you are a adult")
```

```python
# Even or odd using nested if condition
num = int(input("Enter the number"))
if num > 0:
    print("The number is positive")
    if num%2==0:
        print("The number is even")
    else:
        print("The number is odd")
else:
    print("The number is zero or negitive number")       
```

```python
# Nested if condition leap year
year = int(input("Enter the year"))
if year%4==0:
    if year%100==0:
        if year%400==0:
            print(year+ " is a leap year")
        else:
            print(year,  " : is not a leap year")
    else:
        print(year, ": is a leap year")        
else:
    print(year,  ": is not a leap year")       
```
