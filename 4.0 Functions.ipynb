```python
def even_odd(num):
    if num%2 == 0:
        print("The given number is even")
    else:
        print("The given number is odd")
        
even_odd(10)
```

```python
## function with multiple parameters
def multiple_Par(a,b):
    c = a+b
    return c

multiple_Par(4,5)
```

```python
def sub(a,b):
    return a-b

sub(10,6)
```

```python
## Default paramter
def greet(name):
    print(f"Hello {name} welcome to my house")

greet('Pojitha')
```

```python
def greet(name = "Guest"):
    print(f"Hello {'Guest'} welocme to my house")

greet('friend')
```

```python
## Variable length arguments
## Positional arguments and Keyword arguments
#Positional arguments
def print_number(*args):
    for number in args:
        print(number)
        
print_number(1,2,3,4,5,6,7,"Eswar")
```

```python
## Keyword arguments
def print_values(**kwargs):
    for key,values in kwargs.items():
        print(f"{key}:{values}")
        
print_values(name='Eswar',Age= "25",Country="India")
```

```python
## combing both postion and keyword arguments
def print_details(*args,**kwargs):
    for number in args:
        print(f" postion arguments : {number}")
    for key,value in kwargs.items():
        print(f"{key}:{value}")
        
print_details(1,2,3,4,5,6,"Eswar",name="Surendra",age = "24",Work = "Software Engineer")
```
