```python
# Right aligned triangle
n = 5
for i in range(1,n+1):
    print(" " * (n-1) + "*" * i)
```

```python
# Pyramid  pattren
n = 5
for i in range(1, n+1):
    print(" " * (n-1) + "*" * (2*i-1))
```

```python
# Diamond shape
for i in range(1, n+1):
    print(" " * (n - i) + "*" * (2*i - 1))
# Lower part
for i in range(n-1, 0, -1):
    print(" " * (n - i) + "*" * (2*i - 1))
```

```python
# Factorial using while loop
n = int(input("Enter a fact number"))
factorial = 1
i=1
while i <= n:
    factorial = factorial*i
    i += 1
print(f"The factorial of {n} is {factorial}")
    
```

```python
# Calculate the sum of the digts
n = int(input("Enter a number"))
sum_of_digits = 0
while n > 0:
    digit = n % 10
    sum_of_digits += digit
    n = n // 10
print(f"The sum of the digits is {sum_of_digits}")
```

List() methods
1.append() - Adding an elemnt of the end
2.insert() - Replace the item element postion and add that postion
3.remove() - Removing the first occurance of the element
4.pop()  - Remove the return of the last element
5.index() - Finding the element index postion
6.sort() - sort the list in asscending order

```python
# List Methos
fruits = ["Apple","Orange","banana","Cherry","Kiwi"]
fruits.append("Orange")
print(fruits)
fruits.insert(1,"guva")
print(fruits)
fruits.remove("banana")
print(fruits)
fruits.pop()
print(fruits)
index = fruits.index("guva")
print(index)
print(fruits.sort())
print(fruits.reverse())


```
