Conditional statements
1. if statement
2. elif statement
3. nested if statement

```python
## if statement
age=18
if age >= 18:
    print("You are a adult")
```

```python
## else statement
age = 17
if age >= 18:
    print('You are adult')
else:
    print('You are a teenager')
```

```python
## elif statement
age = 20
if age < 13:
    print('you are child')
elif age < 18:
    print('You are teenager')
else:
    print('you are adult')
```

```python
## Nested if Statement
## number even or odd number
num = int(input("Enter the number"))
if num > 0:
    print('The number is positive')
    if num%2 == 0:
        print('The number is even')
    else:
        print('The number is odd')
else:
    print('The number is zero or negative')
```

```python
## Examples for conditional statements
## using nested if leap year
year = int(input('Enter the year'))
if year%4==0:
    if year%100==0:
        if year%400==0:
            print(year, ' is a leap year')
        else:
            print(year, " is not leap year")
    else:
        print(year, " is a leap year")
else:
    print(year, " is not a leap year")
            
```

```python
## simple calculator
num1 = float(input('Enter the first number'))
num2 = float(input('Enter the second number'))
operator  = input('Enter  operators  (+,-,*,/) : ')

if operator == '+':
    result = num1 + num2
elif operator == '-':
    result = num1 - num2
elif operator == '*':
    result = num1 * num2
elif operator == '/':
    if num2 != 0:
        result = num1 / num2
    else:
        result = 'Error division by zero'
else:
    print("Invalid operation")
    
print(result, 'result')

```

```python
## Determine the ticket price based on the age
## Whether the person is student 

age = int(input('Enter the age'))
is_student = input('Are you a student (yes/no) : ').lower
if age < 5:
    price = 'Free'
elif age <= 12:
    price = '5$'
elif age <= 17:
    if is_student == 'yes':
        price = '12$'
    else:
        price = '15$'
elif age <= 64:
    if is_student == 'yes':
        price = '18$'
    else:
        price = '20$'
        
print("Ticket price : ",price)
```
