```python
# Tuple Methos 
# There are 2 methos  
# 1. count()
# 2 . index()

numbers = (1,2,3,4,5)
print(numbers.count(1))
print(numbers.index(3))
```

```python
tuple = (4,4,5,5,6,7)
print(tuple)
```

```python

## packing and unpacking tuple
# 1. Packing the tuple
packed_tuples = 1,"Hello",3.14,True
print(packed_tuples)

# 2. unpacking tuple
a,b,c,d = packed_tuples
print(a)
print(b)
print(c)
print(d)

## unpacking with *

numbers = (1,2,3,4,5,6)
first,*middle,last = numbers
print(first)
print(middle)
print(last)
```

```python
## Nested list
list = [[1,2,3,4],[1,"hello",3.14,True],['a','b','c','d']]
print(list[0:])
print(list[1:3])
print(list[::2])
print(list[::-1])
print(list[0][0:4])
```

```python
## Nested tuples
nested_tuples = ((1,2,3,4),('a','b','c','d'),(1,"hello",3.14,True))
print(nested_tuples[0][0:3])
print(nested_tuples[0:])
print(nested_tuples[1][0:4])
```

```python
## Iterating over the tuple

for sub_tuple in nested_tuples:
    for item in sub_tuple:
        print(item, end = " ")
    print()    
```

```python
for name in nested_tuples:
    for ans in  name:
        print(ans, end = " ")
    print()
```

Conculsion 

Tuples are versatile and useful in many recent world scenarios where an immutable and ordered collection of items is required . They are commonly and return values and as dictionary keys. Understanding how to levarage tuples effectively can improve the efficiency and readability of your python code
