```python
# Arithmetic operator
a = 10 
b = 20 
print(a+b) # Addition
print(a-b) # Substraction
print(a*b) # Multiuplication
print(a/b) # Division
print(a//b) # Floor division
print(a%b) # modulus
print(a**b) # Exponetital
```
