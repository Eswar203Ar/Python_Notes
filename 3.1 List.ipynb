```python
## List Comphresnsion
lst = []
for i in range(10):
    lst.append(i**2)

print(lst)
```

```python
## List comphrension
[x**2 for x in range(10)]
```

```python
## Basic list comphrension 
## Syntax [expression for item in iterable]
square = [x**2 for x in range(10)]
print(square)
```

```python
## List comphrension with condition
## [expression for item in iterable  if conditioion]
even_number = [num  for num in range(20) if num%2==0]
print(even_number)
```

```python
## Without list comphrension expression
lst = []
for num in range(30):
    if num%2 == 0:
        lst.append(num)
            
print(lst)
```

```python
## Nested list comphrension
## Syntax [expression for item 1 in iterable for item2 in iterable ]
list1 = [1,2,3,4,5,6]
list2 = ['a','b','c','d','E']
pair = [[i,j] for i in list1 for j in list2]
print(pair)
```

```python
## List comphrension with function'
words = ["Hello",'world','Python','List','Comphrension']
length = [len(word) for word in words]
print(length)
```
