```python
# why function
# function are using reusing the code
## Without using function
num = 24
if num%2 == 0:
    print("The number is even")
else:
    print("The  number is odd")
```

```python
## using functions
def even_or_odd(num):
    if num%2==0:
        print("The number is even")
    else:
        print("The number is odd")
```

```python
## calling the def function
even_or_odd(10)
```

```python
def String_name(name):
    print(f"Welcome {name}")
    
String_name("Eswar")
```

```python
## Function with multiple parameters
def add(a,b):
    c = a+b
    return c
add(10,2)

def sub(a,b):
    return a-b
result = sub(5,3)
print(result)
```

```python
def multiplication(a,b):
    c = a*b
    return c
multiplication(5,3)

def multi(a,b):
    return a*b
result = multi(4,5)
print(result)
```

```python
## Defalut parameters
def greet(name = "Guest"):
    print(f"Hello {name} welcome to sports club")
greet() ## default function
greet("ESwar")
```

```python
## variable length arguments 
## position arguments and keyword arguments
##def print_number(num1,num2,num3,num4):
# Without writting the all paramters simple using to *args
def print_numbers(*args):
    for num in args:
        print(num)

print_numbers(1,2,3,4,5,"Eswar")
```

```python
## Keyword arguments
## keyword arguments are decelared by **Kwargs
## keyword argyments can be key value pairs
def keywor_args(**kwargs):
    for key,value in kwargs.items():
        print(f"{key} : {value}")
keywor_args(name = "Eswar",age = 25,address = "Hyderabad")
```

```python
## combining both postion arguments and keyword arguments
def combination_aruguments(*args,**kwargs):
    for val in args:
        print(f"Position arguments : {val}")
    for key,value in kwargs.items():
        print(f"{key} : {value}")
combination_aruguments(1,2,3,4,"Eswar",name = "Kalyan", age = "26",address = "hyderabad")
```
