Dictinaries

Dictionary are the unordered collection of items. They store 
data in key-value pair . key must be unique and mutable (eg: string, numbers or tuples) while values can be of any type.

```python
# creating the dictinary
empty_dict = {}
print(type(empty_dict))

## another way of creating dictinary
empty_dict = dict()
print(type(empty_dict))
```

```python
student = {"name" : "Eswar", "age" : "25" ,"grade" : "A"}
print(student)
```

```python
## Error 
student = {"name" : "Eswar", "age" : "24", "name" : "Ardhani"}
print(student)

#key value pair must be unique they don't provide error the value should be re-assigned
```

```python
## Accessing dictinoary
student = {"name" : "Eswar", "age" : "25" ,"grade" : "A"}
print(student)

print(student)
print(student['name'])
print(student['age'])
print(student['grade'])
```

```python
## Accessing using get() method
print(student.get('name'))
print(student.get('age'))
print(student.get('grade'))
```

```python
## Modifying the elements
# Dictinary are mutable they can add,update,and delete
student['age'] = 33
print(student)

student['address'] = "India"
print(student)

## delete the key and values pairs

del student['grade']
print(student)
```

```python
## Dictinary methods
# keys : get all the keys
keys = student.keys()
print(keys)

# Get : get all values
values = student.values()
print(values)

# items  get all the keys and values
items = student.items()
print(items)
```

```python
## Shallow copy
student_copy = student
print(student)
print(student_copy)
student['name'] = 'krish'
print(student)
print(student_copy)

## SHALLOW COPY
student_copy = student.copy()
print(student_copy)
print(student)
student['name'] = 'Ardhani'
print(student_copy)
print(student)

## imapct only one copy by using shallow copy
```

```python
## iterating over Dictionaries
# we can use loops to iterate over dictinaries, keys, values,items

for keys in student.keys():
    print(keys)
    
for keys in student.values():
    print(values)
    
for keys, values in student.items():
    print(f"{keys} : {values}")
```

```python
## Iterating over dictinaries
for keys in student.keys():
    print(keys)
    
for keys in student.values():
    print(values)
    
for keys,values in student.items():
    print(f"{keys} : {values}")
```

```python
## Nested Dictinary

student = {
          "student1" : {"name":"Eswar", "age" : 25},
          "student2"  :{"name":"peter", "age" : 32}
}

print(student)
```

```python
## Accessiong the nested Dictinary
print(student['student1']['name'])
print(student['student2']['age'])
```

```python
print(student['student2']['age'])
print(student['student2']['name'])
```

```python
## Dictonary comphrension

square = {x : x**2 for x in range(5)}
print(square)
```

```python
## conditional dictionary statements

evens = {x : x**2 for x in range(10) if x%2 == 0}
print(evens)
```

```python
## praticle examples
# use a dictionary to count frequency of elements in list
numbers = {1,2,2,4,4,4,4,5,6,7,8,9,9,9}
frequency = {}
for number in numbers:
    if number in frequency:
        frequency[number] += 1
    else:
        frequency[number] = 1
print(frequency)
        
```

```python
## merge two dictinaries into one

dict1 = {'a' : 1, 'b' : 2}
dict2 = {'c' : 3, 'd' : 4}
merge_dict = {**dict1, **dict2}
print(merge_dict)
```
